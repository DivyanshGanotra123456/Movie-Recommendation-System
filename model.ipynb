```python
data=pd.merge(ratings,movies,on="movieId")
```

```python
user_item_matrix=data.pivot_table(index='userId',columns='title',values='rating')
```

```python
user_item_matrix
```

```python
user_item_matrix_filled=user_item_matrix.fillna(0)
```

```python
user_item_matrix_filled
```

```python
user_item_matrix.shape
```

```python
movie_matrix=user_item_matrix_filled.T
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(movie_matrix)
```

```python
similarity_df=pd.DataFrame(similarity,index=movie_matrix.index,columns=movie_matrix.index)
```

```python
similarity_df
```

```python
def recommend_systems(movie_name,n=5):
    similar_movies=similarity_df[movie_name].sort_values(ascending=False)
    return similar_movies.iloc[1:n+1]
```

```python
recommend_systems('Toy Story (1995)')
```

```python
from scipy.sparse.linalg import svds
```

```python
matrix=user_item_matrix_filled.values
```

```python
u,sigma,vt=svds(matrix,k=50)
```

```python
import numpy as np
```

```python
sigma=np.diag(sigma)
```

```python
predicted_ratings=np.dot(np.dot(u,sigma),vt)
```

```python
pred_df=pd.DataFrame(predicted_ratings,columns=user_item_matrix_filled.columns,index=user_item_matrix_filled.index)
```

```python
pred_df
```

```python
def get_watched_movies(user_id):
    user_data=user_item_matrix.loc[user_id]
    watched=user_data[user_data>0].index.tolist()
    return watched
```

```python
def recommend_movies_svd(user_id,n=10):
    user_predictions=pred_df.loc[user_id]
    watched_movies=get_watched_movies(user_id)
    recommendations=user_predictions.drop(watched_movies)
    recommendations=recommendations.sort_values(ascending=False)
    return recommendations.head(n)
```

```python
recommend_movies_svd(1)
```
